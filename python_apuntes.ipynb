# Variables

```python
# Las variables son espacios en memoria donde podemos almacenar datos.
# Se crean al asignar un valor a un nombre: nombre_variable = valor

x = 5
y = "Hola"
z = 3.14
print(x)
print(y)
print(z)

# =========================================================
# FUNCIONES INCORPORADAS
# =========================================================
# En Python, tenemos muchas funciones integradas.
# Estas funciones están disponibles globalmente, lo que significa que puedes usarlas sin importar ni configurar.
# Algunas de las funciones integradas más comunes son: print(), len(), type(), int(), float(), str(), input(), list(), dict(), min(), max(), sum(), sorted(), open(), file(), help() y dir()

# Operaciones numéricas
print(abs(-5))        # 5
print(divmod(9,2))    # (4,1)
print(pow(2,3))       # 8
print(round(3.6))     # 4
print(sum([1,2,3]))   # 6
print(min([2,5,1]))   # 1
print(max([2,5,1]))   # 5
print(bin(5))          # '0b101'
print(oct(8))          # '0o10'
print(hex(255))        # '0xff'
print(complex(3,4))    # (3+4j)

# Secuencias, listas y colecciones
print(len("Hola"))                    # 4
print(list((1,2,3)))                  # [1,2,3]
print(tuple([1,2,3]))                 # (1,2,3)
print(set([1,2,2]))                   # {1,2}
print(frozenset([1,2,2]))             # frozenset({1,2})
print(sorted([3,1,2]))                # [1,2,3]
print(list(reversed([1,2,3])))        # [3,2,1]
print(list(enumerate(['a','b'])))    # [(0,'a'),(1,'b')]
print(list(zip([1,2],[3,4])))        # [(1,3),(2,4)]
print(list(range(3)))                 # [0,1,2]
print(list(filter(lambda x: x>2,[1,2,3]))) # [3]
print(list(map(str,[1,2])))           # ['1','2']
print(any([0,1]))                      # True
print(all([1,1]))                      # True

# Texto y tipos básicos
print(str(5))      # '5'
print(chr(65))     # 'A'
print(ord('A'))    # 65
print(ascii("ñ"))  # "'\\xf1'"
print("{} + {} = {}".format(2,3,5))  # '2 + 3 = 5'

# Clases, objetos y atributos
print(isinstance(5,int))  # True
# get/set/hasattr/delattr se usan en objetos:
class Persona:
    def __init__(self,nombre):
        self.nombre = nombre
p = Persona("Ana")
print(getattr(p,"nombre"))  # Ana
setattr(p,"edad",30)
print(hasattr(p,"edad"))    # True
delattr(p,"edad")
print(hasattr(p,"edad"))    # False

# Entrada, salida y entorno
# print(), input() ya visto
# open() abre archivos, help() muestra ayuda, dir() lista atributos/métodos
print(globals())  # Variables globales
print(locals())   # Variables locales
print(vars(p))    # Diccionario del objeto

# Control y ejecución
print(eval('3+2'))   # 5
exec('x=5')
print(x)             # 5
print(callable(print))  # True

# Tipos y conversión
print(bool(0))         # False
print(bytes("hola","utf-8")) # b'hola'
print(bytearray("hola","utf-8")) # bytearray(b'hola')
print(memoryview(b"hola"))      # <memory at ...>
print(type(3))          # <class 'int'>
print(id(42))           # id único
print(hash("hola"))     # hash del string
print(repr("texto"))    # "'texto'"

# Otros
print(slice(1,3))         # slice(1,3)
it = iter([1,2,3])
print(next(it))            # 1
print(__import__('math'))  # importa math
```

# Operadores

```python
# =========================================================
# OPERADORES DE ASIGNACIÓN
# =========================================================
a = 5
b = 3
a += b   # equivalente a a = a + b
a -= b   # a = a - b
a *= b   # a = a * b
a /= b   # a = a / b
a %= b   # a = a % b
a //= b  # a = a // b
a **= b  # a = a ** b

# =========================================================
# OPERADORES ARITMÉTICOS
# =========================================================
print(5 + 3)   # Suma
print(5 - 3)   # Resta
print(5 * 3)   # Multiplicación
print(5 / 3)   # División
print(5 % 3)   # Módulo
print(5 // 3)  # División entera
print(5 ** 3)  # Exponente

# =========================================================
# OPERADORES DE COMPARACIÓN
# =========================================================
print(5 == 3)  # Igual
print(5 != 3)  # Distinto
print(5 > 3)   # Mayor que
print(5 < 3)   # Menor que
print(5 >= 3)  # Mayor o igual
print(5 <= 3)  # Menor o igual



# Tip visual en VS Code con Fira Code:
# ==   → igual
# <=   → menor o igual (≤)


# Operadores lógicos
a = True
b = False
print(a and b)   # AND → ambos True → False
print(a or b)    # OR → uno True → True
print(not a)     # NOT → invierte → False


# Tip rápido:
# AND → todos deben ser True
# OR → basta con uno True
# NOT → invierte el valor lógico

| A     | B     | A and B | A or B | not A |
| ----- | ----- | ------- | ------ | ----- |
| True  | True  | True    | True   | False |
| True  | False | False   | True   | False |
| False | True  | False   | True   | True  |
| False | False | False   | False  | True  |
```

# Strings

```python
# a) Secuencias de escape en cadenas
# En Python y otros lenguajes de programación, la secuencia de escape se denomina "\" seguida de un carácter.
# Veamos los caracteres de escape más comunes:
# \n: nueva línea
# \t: Tabulador significa (8 espacios)
# \\: Barra invertida
# \': Comilla simple (')
# \": Comillas dobles (")

# Ejemplo: nueva línea
print('I hope everyone is enjoying the Python Challenge.\nAre you ?')
# I hope every one is enjoying the Python Challenge.
# Are you ?

# Si incluimos un “\” antes de la secuencia la anula.
print('I hope everyone is enjoying the Python Challenge.\\nAre you ?')
# I hope every one is enjoying the Python Challenge. \nAre you ?

# =========================================================
# b) Formato de cadenas
# En Python existen muchas maneras de formatear cadenas.
# El operador "%" se utiliza para formatear un conjunto de variables dentro de una "tupla" (una lista de tamaño fijo),
# junto con una cadena de formato que contiene texto normal junto con "especificadores de argumentos", símbolos especiales como "%s", "%d", "%f" y "%.number of digitsf".
# %s - Cadena (o cualquier objeto con una representación de cadena, como números)
# %d - Números enteros
# %f - Números de punto flotante
# "%.number of digitsf": números de punto flotante con precisión fija

name, surname, age = "Brais", "Moure", 35

# Versiones de formateo
print("Mi nombre es {} {} y mi edad es {}".format(name, surname, age))  # Versión Óptima
print("Mi nombre es %s %s y mi edad es %d" % (name, surname, age))      # Versión Óptima Antigua
print("Mi nombre es " + name + " " + surname + " y mi edad es " + str(age))
print(f"Mi nombre es {name} {surname} y mi edad es {age}")               # Versión Más Óptima

# =========================================================
# c) Cadenas de Python como secuencias de caracteres
# Las cadenas de Python son secuencias de caracteres y comparten sus métodos básicos de acceso con otras secuencias ordenadas de objetos de Python: listas y tuplas.
# La forma más sencilla de extraer caracteres individuales de cadenas (y miembros individuales de cualquier secuencia) es descomprimirlos en las variables correspondientes.

# c.I.) Desempacando personajes
language = 'Python'
a,b,c,d,e,f = language # unpacking sequence characters into variables
print(a) # P
print(b) # y
print(c) # t
print(d) # h
print(e) # o
print(f) # n

# Otro ejemplo:
language = "raton"
a, b, c, d, e = language
print(a) # r
print(e) # n

# c.II.) Acceso a caracteres en cadenas por índice
# En programación, el conteo empieza desde cero. Por lo tanto, la primera letra de una cadena está en el índice cero y la última letra equivale a la longitud de la cadena menos uno. Si contamos desde el final empieza por -1 (no existe el -0)
# ['P', 'y', 't', 'h', 'o', 'n']
#  0    1    2    3    4    5

language = "Python"
language_slice = language[1:3]
print(language_slice)    # yt (empieza desde el índice 1 y llega hasta el índice 3, sin incluirlo)

language_slice = language[1:]
print(language_slice)    # ython

language_slice = language[-2]
print(language_slice)    # o

language_slice = language[0:6:2]
print(language_slice)    # Pto (Empieza en el índice 0 y va saltando de 2 en 2 hasta antes del índice 6)

reversed_language = language[::-1]
print(reversed_language) # nohtyP

# =========================================================
# 1️⃣ Formateo e inserción
# str.format() → Inserta valores en {} → "Hola {}".format("Ana") → "Hola Ana"
# F-strings (f"") → Inserta variables directamente → nombre="Ana"; f"Hola {nombre}" → "Hola Ana"
# %-formatting → Inserción estilo C → "Edad: %d" % 30 → "Edad: 30"

nombre = "Ana"
print("Hola {}".format(nombre))
print(f"Hola {nombre}")
print("Edad: %d" % 30)

# =========================================================
# 2️⃣ Mayúsculas, minúsculas y capitalización
print("hola mundo".upper())      # HOLA MUNDO
print("HOLA MUNDO".lower())      # hola mundo
print("hola mundo".capitalize()) # Hola mundo
print("hola mundo".title())      # Hola Mundo
print("Hola Mundo".swapcase())   # hOLA mUNDO
print("ß".casefold())            # ss

# =========================================================
# 3️⃣ Alineación y relleno
print("hi".center(10))           # '   hi    '
print("hi".ljust(10))            # 'hi        '
print("hi".rjust(10))            # '        hi'
print("42".zfill(5))             # '00042'
print("hi".center(6,'_'))        # '__hi__'

# =========================================================
# 4️⃣ Eliminación de espacios y caracteres
print(" hola ".strip())           # 'hola'
print(" hola".lstrip())           # 'hola'
print("hola ".rstrip())           # 'hola'
print("hola".replace("o","a"))    # 'hala'
print("Python".removeprefix("Py"))# 'thon'
print("hello.py".removesuffix(".py")) # 'hello'

# =========================================================
# 5️⃣ Búsqueda y posición
print("hola mundo".find("o"))    # 1
print("hola mundo".rfind("o"))   # 7
print("hola mundo".index("o"))   # 1
print("hola mundo".rindex("o"))  # 7
print("hola mundo".startswith("ho")) # True
print("hola mundo".endswith("do"))   # True
print("banana".count("a"))           # 3
print("a-b-c".partition("-"))        # ('a','-','b-c')
print("a-b-c".rpartition("-"))       # ('a-b','-','c')

# =========================================================
# 6️⃣ División y unión
print("a,b,c".split(","))           # ['a','b','c']
print("a,b,c".rsplit(",",1))        # ['a,b','c']
print("a\nb\nc".splitlines())       # ['a','b','c']
print(",".join(['a','b','c']))      # "a,b,c"

# =========================================================
# 7️⃣ Verificación de contenido
print("abc123".isalnum()) → True si solo letras o números → "abc123".isalnum() → True
print("abc".isalpha()) → True si solo letras → "abc".isalpha() → True
print("123".isdigit()) → True si solo dígitos → "123".isdigit() → True
print("²34".isnumeric()) → True si caracteres numéricos → "²34".isnumeric() → True
print("123".isdecimal()) → True si números decimales → "123".isdecimal() → True
print(" ".isspace()) → True si solo espacios → " ".isspace() → True
print("hola".islower()) → True si toda la cadena está en minúsculas → "hola".islower() → True
print("HOLA".isupper()) → True si toda la cadena está en mayúsculas → "HOLA".isupper() → True
print("Hola Mundo".istitle()) → True si cada palabra empieza con mayúscula → "Hola Mundo".istitle() → True
print("Hola".isascii()) → True si todos los caracteres son ASCII → "Hola".isascii() → True
print("var_1".isidentifier()) → True si puede ser identificador válido → "var_1".isidentifier() → True

# =========================================================
# 8️⃣ Transformación avanzada
print("a\tb".expandtabs(4))   # 'a    b'
print("hola".encode())        # b'hola'
tabla=str.maketrans("aeiou","12345")
print("hola".translate(tabla)) # 'h4l1'

# =========================================================
# 9️⃣ Extra
print(len("hola"))             # 4
```

```python
# =========================================================
# ACCESO A MÉTODOS Y ATRIBUTOS MEDIANTE EL PUNTO (.)
# =========================================================
# En Python, el punto (.) se usa para acceder a algo que “pertenece” a un objeto.
# Cada tipo de dato (listas, cadenas, números, etc.) tiene sus propios métodos y atributos.
# Cuando escribes el nombre del objeto seguido de un punto, tu editor o entorno (como Colab o VS Code)
# te mostrará una lista de sugerencias de los métodos disponibles para ese tipo de objeto.

# ---------------------------------------------------------
# Ejemplo con listas:
# ---------------------------------------------------------
my_other_list = [1, 2, 3]

# Si escribes en tu editor:
# my_other_list.
# verás sugerencias como:
# append() → Añade un elemento al final de la lista
# clear()  → Vacía la lista completamente
# pop()    → Elimina un elemento (por índice o el último por defecto)
# sort()   → Ordena los elementos
# count()  → Cuenta cuántas veces aparece un valor

# Ejemplos prácticos:
my_other_list.append(4)
print("Después de append:", my_other_list)  # [1, 2, 3, 4]

my_other_list.pop()
print("Después de pop:", my_other_list)     # [1, 2, 3]

my_other_list.sort(reverse=True)
print("Después de sort descendente:", my_other_list)  # [3, 2, 1]

print("Ocurrencias del número 2:", my_other_list.count(2))  # 1

my_other_list.clear()
print("Después de clear:", my_other_list)  # []

# ---------------------------------------------------------
# Ejemplo con números:
# ---------------------------------------------------------
# Si el tipo de dato fuera un número (int o float),
# las sugerencias serían diferentes, por ejemplo:
# round() — Redondea un número con n decimales.
# Ejemplo:

numero = 3.14159
print("Número redondeado a 2 decimales:", round(numero, 2))  # 3.14
```

# Colecciones_de_datos

```python
# ------------------------------------------------------------
# COMPARATIVA COMPLETA DE COLECCIONES EN PYTHON
# ------------------------------------------------------------

# Tipo de colección | Sintaxis | Ordenada | Mutable | Duplicados | Acceso | Métodos comunes

# Lista
# [1,2,3]
# Ordenada: Sí
# Mutable: Sí
# Duplicados: Sí
# Acceso: Por índice
# Métodos comunes: append(), insert(), remove(), pop(), clear(), sort(), reverse(), copy(), count(), index()

# Tupla
# (1,2,3)
# Ordenada: Sí
# Mutable: No
# Duplicados: Sí
# Acceso: Por índice
# Métodos comunes: count(), index()

# Set
# {1,2,3}
# Ordenada: No
# Mutable: Sí
# Duplicados: No
# Acceso: No (solo por iteración)
# Métodos comunes: add(), remove(), discard(), pop(), clear(), union(), intersection(), difference(), issubset(), issuperset()

# Diccionario
# {'clave': valor}
# Ordenada: Sí (desde Python 3.7+)
# Mutable: Sí
# Duplicados: No (por clave)
# Acceso: Por clave
# Métodos comunes: keys(), values(), items(), get(), pop(), update(), clear(), copy()

# ------------------------------------------------------------
# CUÁNDO USAR CADA TIPO DE COLECCIÓN EN PYTHON
# ------------------------------------------------------------

# 1️⃣ LISTA [ ]
# Cuando necesitas modificar los elementos (agregar, eliminar, cambiar).
# Cuando el orden importa y quieres acceder por índice.
# Cuando quieres permitir duplicados.
# Para colecciones dinámicas que cambian en tiempo de ejecución.

frutas = ['manzana', 'banana']
frutas.append('naranja')
print(frutas)
# ['manzana', 'banana', 'naranja']

# ------------------------------------------------------------

# 2️⃣ TUPLA ( )
# Cuando los datos no deben cambiar (constantes).
# Para retornar múltiples valores de una función.
# Cuando necesitas inmutabilidad por seguridad.
# Cuando los elementos deben ser hashables (por ejemplo, para usar como clave en diccionarios o sets).

coordenadas = (10.0, 20.0)
print(coordenadas)
# (10.0, 20.0)
# No se puede modificar: coordenadas[0] = 5  ❌ (Error)

# ------------------------------------------------------------

# 3️⃣ SET { }
# Cuando no quieres duplicados.
# Para operaciones de conjunto (unión, intersección, diferencia).
# Cuando el orden no importa.
# Cuando quieres acceso rápido para verificar existencia (x in set).

numeros = {1, 2, 2, 3}
print(numeros)
# {1, 2, 3}
numeros.add(4)
print(numeros)
# {1, 2, 3, 4}

# ------------------------------------------------------------

# 4️⃣ DICCIONARIO {clave: valor}
# Cuando quieres asociar valores a claves únicas.
# Para acceso rápido por clave en lugar de índice.
# Cuando los datos pueden cambiar, pero la clave identifica de manera única cada valor.

persona = {'nombre': 'Ana', 'edad': 30}
persona['edad'] = 31
print(persona['nombre'])
# Ana
print(persona)
# {'nombre': 'Ana', 'edad': 31}
```

## Lists

```python
# ------------------------------------------------------------
# TIPOS DE DATOS DE COLECCIÓN EN PYTHON
# ------------------------------------------------------------
# Lista (list): colección ordenada y modificable. Permite duplicados.
# Tupla (tuple): colección ordenada e inmutable. Permite duplicados.
# Set (set): colección desordenada, no indexada y no modificable,
#            aunque se pueden añadir nuevos elementos. No permite duplicados.
# Diccionario (dict): colección desordenada, modificable e indexada. Sin duplicados.

# ------------------------------------------------------------
# LISTAS
# ------------------------------------------------------------
# Una lista es una colección de diferentes tipos de datos, ordenada y modificable (mutable).
# Puede estar vacía o contener elementos de diferentes tipos.

# Crear listas
my_list = list()
my_other_list = []

# Ejemplo con frutas
fruits = ['banana', 'orange', 'mango', 'lemon']
print('Fruits:', fruits)
print('Number of fruits:', len(fruits))
# Fruits: ['banana', 'orange', 'mango', 'lemon']
# Number of fruits: 4

# Acceso por índice
second_fruit = fruits[1]
print(second_fruit)
# orange

# Índice negativo (empieza desde el final)
second_last = fruits[-2]
print(second_last)
# mango

# Desempaquetado de lista
fruits = ['banana', 'orange', 'mango', 'lemon', 'lime', 'apple']
first_fruit, second_fruit, third_fruit, *rest = fruits
print(first_fruit)
# banana

# ------------------------------------------------------------
# MODIFICACIÓN DE LISTAS
# ------------------------------------------------------------
fruits = ['banana', 'orange', 'mango', 'lemon']

# Reemplazar elementos por índice
fruits[0] = 'avocado'
print(fruits)
# ['avocado', 'orange', 'mango', 'lemon']

fruits[1] = 'apple'
print(fruits)
# ['avocado', 'apple', 'mango', 'lemon']

# Reemplazar el último elemento
last_index = len(fruits) - 1
fruits[last_index] = 'lime'
print(fruits)
# ['avocado', 'apple', 'mango', 'lime']

# ------------------------------------------------------------
# MÉTODOS DE LISTAS
# ------------------------------------------------------------

# Agregar elementos al final
lista = [1, 2]
lista.append(3)
print(lista)
# [1, 2, 3]

# Insertar elementos en un índice concreto
lista = [1, 3]
lista.insert(1, 2)
print(lista)
# [1, 2, 3]

# Eliminar un elemento por valor
lista = [1, 2, 3]
lista.remove(2)
print(lista)
# [1, 3]

# Eliminar elemento por índice y devolverlo
lista = [1, 2, 3]
primer_elemento = lista.pop(0)
print(primer_elemento)
# 1
print(lista)
# [2, 3]

# Eliminar todos los elementos
lista = [1, 2, 3]
lista.clear()
print(lista)
# []

# Reemplazar elemento por índice
lista = [1, 2, 3]
lista[1] = 10
print(lista)
# [1, 10, 3]

# Buscar índice de un elemento
lista = [10, 20, 30]
print(lista.index(20))
# 1

# Contar cuántas veces aparece un elemento
lista = [1, 2, 2, 3]
print(lista.count(2))
# 2

# Comprobar si un elemento está en la lista
lista = [1, 2, 3]
print(2 in lista)
# True

# Ordenar lista ascendente
lista = [3, 1, 2]
lista.sort()
print(lista)
# [1, 2, 3]

# Ordenar descendente
lista = [3, 1, 2]
lista.sort(reverse=True)
print(lista)
# [3, 2, 1]

# Invertir lista
lista = [1, 2, 3]
lista.reverse()
print(lista)
# [3, 2, 1]
```

## Tuplas

```python
# ------------------------------------------------------------
# TUPLAS EN PYTHON
# ------------------------------------------------------------
# Una tupla es una colección de diferentes tipos de datos, ordenada e inmutable.
# Se escriben entre paréntesis ().
# Una vez creada, no se pueden modificar sus valores.
# No se pueden usar métodos de adición, inserción o eliminación en una tupla, ya que no es modificable (inmutable).
# A diferencia de una lista, una tupla tiene pocos métodos.

# ------------------------------------------------------------
# MÉTODOS DE TUPLAS
# ------------------------------------------------------------
# tuple(): crear una tupla vacía
# count(): contar el número de un elemento especificado en una tupla
# index(): encontrar el índice de un elemento específico en una tupla
# operador + : unir dos o más tuplas y crear una nueva tupla

# Crear tuplas vacías
my_tuple = tuple()
my_other_tuple = ()
print(type(my_tuple))
# <class 'tuple'>
print(type(my_other_tuple))
# <class 'tuple'>

# ------------------------------------------------------------
# COMPARACIÓN CON LISTA
# ------------------------------------------------------------
# Mutabilidad
# Lista: mutable → puedes cambiar elementos
lista = [1, 2, 3]
lista[0] = 10
print(lista)
# [10, 2, 3]
lista.append(4)
print(lista)
# [10, 2, 3, 4]

# Tupla: inmutable → no puedes cambiar elementos
tupla = (1, 2, 3)
# tupla[0] = 10  # ❌ Error: 'tuple' object does not support item assignment
# tupla.append(4)  # ❌ Error: 'tuple' object has no attribute 'append'

# ------------------------------------------------------------
# ACCESO A ELEMENTOS
# ------------------------------------------------------------
# Ambos permiten acceder a elementos por índice
print(lista[1])
# 2
print(tupla[2])
# 3

# ------------------------------------------------------------
# OPERACIONES VÁLIDAS
# ------------------------------------------------------------
# Slicing (funciona igual en lista y tupla)
print(lista[1:3])
# [2, 3]
print(tupla[0:2])
# (1, 2)

# Longitud
print(len(lista))
# 4
print(len(tupla))
# 3

# Concatenación y repetición
print(lista + [5, 6])
# [10, 2, 3, 4, 5, 6]
print(tupla + (4, 5))
# (1, 2, 3, 4, 5)
print(lista * 2)
# [10, 2, 3, 4, 10, 2, 3, 4]
print(tupla * 2)
# (1, 2, 3, 1, 2, 3)

# ------------------------------------------------------------
# RESUMEN RÁPIDO
# ------------------------------------------------------------
# Lista [ ] → mutable; permite métodos de modificación como append, insert, remove, pop, clear;
# ideal para colecciones que cambian.
# Tupla ( ) → inmutable; no permite métodos de modificación; ideal para datos fijos o constantes.

# ------------------------------------------------------------
# CUÁNDO USAR LISTA O TUPLA
# ------------------------------------------------------------
# ✅ Usar lista [ ] cuando:
# - Necesitas modificar los elementos después de crear la colección.
# - Quieres ordenar o reorganizar los datos.
# - La colección cambia de tamaño dinámicamente.
# - Vas a iterar y actualizar los elementos con frecuencia.

# ✅ Usar tupla ( ) cuando:
# - Los datos no deben cambiar (constantes).
# - Quieres seguridad, evitando modificaciones accidentales.
# - Necesitas que los elementos sean hashables (por ejemplo, para usarlos como clave en un diccionario).
# - Quieres retornar múltiples valores de una función de manera ordenada.
```

```python
# ------------------------------------------------------------
# DICCIONARIOS EN PYTHON
# ------------------------------------------------------------
# Un diccionario es una estructura de datos que almacena pares de clave-valor.
# Cada clave debe ser única y hashable (ej. números, strings, tuplas).
# Cada valor puede ser cualquier tipo de dato: número, string, lista, otro diccionario…

# ------------------------------------------------------------
# CÓMO CREAR UN DICCIONARIO
# ------------------------------------------------------------
# Diccionario vacío
diccionario_vacio = {}
print(diccionario_vacio)
# {}

# Diccionario con elementos
persona = {
    "nombre": "Ana",
    "edad": 25,
    "pais": "España"
}
print(persona)
# {'nombre': 'Ana', 'edad': 25, 'pais': 'España'}

# ------------------------------------------------------------
# ACCEDER A LOS VALORES
# ------------------------------------------------------------
print(persona["nombre"])  # Ana
print(persona["edad"])    # 25

# Usando .get() para no dar error si la clave no existe
print(persona.get("profesion", "No especificado"))  # No especificado

# ------------------------------------------------------------
# MODIFICAR VALORES
# ------------------------------------------------------------
persona["edad"] = 26  # Cambia la edad
persona["profesion"] = "Ingeniera"  # Añade una nueva clave
print(persona)
# {'nombre': 'Ana', 'edad': 26, 'pais': 'España', 'profesion': 'Ingeniera'}

# ------------------------------------------------------------
# ELIMINAR ELEMENTOS
# ------------------------------------------------------------
del persona["pais"]      # Borra la clave 'pais'
print(persona)
# {'nombre': 'Ana', 'edad': 26, 'profesion': 'Ingeniera'}

valor_eliminado = persona.pop("edad")      # Borra y devuelve el valor
print(valor_eliminado)
# 26
print(persona)
# {'nombre': 'Ana', 'profesion': 'Ingeniera'}

# ------------------------------------------------------------
# RECORRER UN DICCIONARIO
# ------------------------------------------------------------
# Recorrer claves
for clave in persona:
    print(clave)
# nombre
# profesion

# Recorrer valores
for valor in persona.values():
    print(valor)
# Ana
# Ingeniera

# Recorrer clave-valor
for clave, valor in persona.items():
    print(clave, ":", valor)
# nombre : Ana
# profesion : Ingeniera

# ------------------------------------------------------------
# MÉTODOS ÚTILES DE DICCIONARIOS
# ------------------------------------------------------------
# .keys() -> Devuelve todas las claves
print(persona.keys())
# dict_keys(['nombre', 'profesion'])

# .values() -> Devuelve todos los valores
print(persona.values())
# dict_values(['Ana', 'Ingeniera'])

# .items() -> Devuelve pares (clave, valor)
print(persona.items())
# dict_items([('nombre', 'Ana'), ('profesion', 'Ingeniera')])

# .get(clave, default) -> Devuelve valor de la clave o default
print(persona.get("edad", "No especificado"))
# No especificado

# .pop(clave) -> Elimina clave y devuelve su valor
persona["edad"] = 30
valor = persona.pop("edad")
print(valor)
# 30

# .update(dict2) -> Añade o actualiza con otro diccionario
otra_persona = {"ciudad": "Madrid", "profesion": "Doctora"}
persona.update(otra_persona)
print(persona)
# {'nombre': 'Ana', 'profesion': 'Doctora', 'ciudad': 'Madrid'}
```

# Conditionals

```python
# ------------------------------------------------------------
# SENTENCIAS CONDICIONALES EN PYTHON
# ------------------------------------------------------------
# De forma predeterminada, las sentencias en scripts de Python
# se ejecutan secuencialmente de arriba a abajo.
# Para modificar el flujo de ejecución podemos usar:
# 1️⃣ Ejecución condicional (if, else, elif)
# 2️⃣ Ejecución repetitiva (while, for) -> se verá más adelante

# ------------------------------------------------------------
# 1. IF
# Se ejecuta un bloque de código si la condición es verdadera.
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
# Solo se ejecuta si la condición es True

# ------------------------------------------------------------
# 2. ELSE
# Se ejecuta si la condición del if es falsa.
edad = 16
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")

# ------------------------------------------------------------
# 3. ELIF (else if)
# Comprueba varias condiciones de manera secuencial.
nota = 85
if nota >= 90:
    print("Sobresaliente")
elif nota >= 70:
    print("Aprobado")
else:
    print("Suspenso")
# Python evalúa de arriba hacia abajo y ejecuta solo la primera condición verdadera

# ------------------------------------------------------------
# 4. OPERADORES LÓGICOS EN CONDICIONALES
# and -> todas las condiciones deben ser verdaderas
# or  -> al menos una condición debe ser verdadera
# not -> niega la condición
edad = 20
tiene_id = True
if edad >= 18 and tiene_id:
    print("Puede entrar")

# ------------------------------------------------------------
# 5. CONDICIONAL SIMPLE EN UNA LÍNEA
edad = 16
print("Mayor" if edad >= 18 else "Menor")
# Útil para resumir condicionales sencillos en una sola línea

# ------------------------------------------------------------
# RESUMEN VISUAL
# Condicional    | Qué hace
# if            -> Ejecuta código si la condición es verdadera
# else          -> Ejecuta código si la condición del if es falsa
# elif          -> Comprueba otra condición si las anteriores fueron falsas
# and, or, not  -> Operadores lógicos para combinar condiciones
```

# Lops

```python
# ------------------------------------------------------------
# BUCLES EN PYTHON (LOOPS)
# ------------------------------------------------------------
# Un bucle sirve para repetir un bloque de código varias veces
# sin tener que escribirlo muchas veces manualmente.

# ------------------------------------------------------------
# 1. FOR -> Recorre elementos de una secuencia
# Se puede usar con listas, cadenas, rangos, etc.

# Ejemplo con range
for i in range(5):   # range(5) genera los números 0,1,2,3,4
    print(i)
# Imprime: 0 1 2 3 4

# Ejemplo con lista
frutas = ["manzana", "pera", "uva"]
for fruta in frutas:
    print(fruta)
# Imprime: manzana pera uva

# ------------------------------------------------------------
# 2. WHILE -> Repite mientras se cumpla una condición
# Ejecuta el bloque mientras la condición sea verdadera

contador = 0
while contador <= 4:
    print("Hola", contador)
    contador += 1   # aumenta el contador
# Imprime: Hola 0 Hola 1 Hola 2 Hola 3 Hola 4

# Ejemplo incrementando de 2 en 2
contador = 0
while contador <= 4:
    print("Hola", contador)
    contador += 2
# Imprime: Hola 0 Hola 2 Hola 4

# ⚠ Cuidado: si no actualizas la condición (contador += 1), el bucle será infinito

# ------------------------------------------------------------
# 3. BREAK y CONTINUE
# break -> rompe el bucle antes de que termine
for i in range(5):
    if i == 3:
        break  # se detiene cuando i = 3
    print(i)
# Imprime: 0 1 2

# continue -> salta a la siguiente iteración sin ejecutar el resto del bloque
for i in range(5):
    if i == 2:
        continue  # salta el número 2
    print(i)
# Imprime: 0 1 3 4

# ------------------------------------------------------------
# 4. FOR ... ELSE
# El else se ejecuta solo si el bucle termina sin usar break
for i in range(3):
    print(i)
else:
    print("Bucle terminado sin interrupciones")
# Imprime: 0 1 2 Bucle terminado sin interrupciones

# ------------------------------------------------------------
# RESUMEN VISUAL
# Tipo      | Qué hace                                  | Cuándo usar
# for       -> Recorre una secuencia (lista, cadena, rango) -> Sabes cuántas veces quieres repetir
# while     -> Repite mientras una condición sea verdadera   -> No sabes cuántas veces se repetirá
# break     -> Sale del bucle                            -> Cuando quieres detener antes de tiempo
# continue  -> Salta a la siguiente iteración           -> Para omitir un caso específico
```

# Functions

```python
# ------------------------------------------------------------
# FUNCIONES EN PYTHON
# ------------------------------------------------------------
# Una función es un bloque de código que se puede reutilizar
# para realizar una tarea específica.
# Sirven para organizar el código, hacerlo más limpio y evitar repetir instrucciones.

# ------------------------------------------------------------
# 1. Cómo se define una función
def saludar():
    print("Hola, ¿cómo estás?")

# Aquí:
# def      -> indica que estás definiendo una función
# saludar  -> nombre de la función
# ()       -> paréntesis (pueden contener parámetros)
# :        -> marca el inicio del bloque de la función
# El código indentado es lo que se ejecuta

# ------------------------------------------------------------
# 2. Cómo se llama (ejecuta) una función
saludar()
saludar()
# Imprime:
# Hola, ¿cómo estás?
# Hola, ¿cómo estás?

# ------------------------------------------------------------
# 3. Funciones con parámetros (entrada)
def saludar_persona(nombre):
    print("Hola", nombre)

saludar_persona("Víctor")
# Imprime: Hola Víctor

# ------------------------------------------------------------
# 4. Funciones que devuelven valores (return)
def sumar(a, b):
    return a + b

resultado = sumar(5, 3)
print(resultado)
# Muestra: 8

# ------------------------------------------------------------
# 5. Funciones con valores por defecto
def saludar(nombre="amigo"):
    print("Hola", nombre)

saludar()           # usa el valor por defecto
saludar("Lucía")    # usa el valor que pasamos
# Imprime:
# Hola amigo
# Hola Lucía

# ------------------------------------------------------------
# 6. Funciones anónimas (lambda)
# Son funciones pequeñas de una sola línea, sin nombre
cuadrado = lambda x: x ** 2
print(cuadrado(4))
# Resultado: 16

# ------------------------------------------------------------
# Diferencias entre def y return

# Versión 1 — usando print (solo muestra, no devuelve)
def hacer_cafe_print():
    print("☕ Tu café está listo!")

hacer_cafe_print()
# Salida:
# ☕ Tu café está listo!

# No se puede guardar el resultado:
cafe = hacer_cafe_print()
print(cafe)
# Resultado:
# ☕ Tu café está listo!
# None

# "None" significa que la función no devolvió nada (solo imprimió)

# Versión 2 — usando return (devuelve el resultado)
def hacer_cafe_return():
    return "☕ Tu café está listo!"

cafe = hacer_cafe_return()
print(cafe)
# Salida:
# ☕ Tu café está listo!

# Se puede usar el resultado más adelante:
print(cafe.upper())
# Resultado:
# ☕ TU CAFÉ ESTÁ LISTO!
```

# Classes

```python
# ------------------------------------------------------------
# CLASES Y OBJETOS EN PYTHON
# ------------------------------------------------------------
# Una clase es como un molde o plantilla para crear objetos.
# Un objeto es una instancia concreta de esa clase, con sus propios valores.
# Clase → la definición, como un plano.
# Objeto → algo creado a partir de la clase, como una casa construida a partir del plano.

# ------------------------------------------------------------
# 1. Cómo definir una clase
class Persona:
    # Constructor (inicializa atributos del objeto)
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    # Método (función dentro de la clase)
    def saludar(self):
        print(f"Hola, mi nombre es {self.nombre} y tengo {self.edad} años")

# Explicación:
# class Persona: → define la clase Persona.
# __init__ → método especial que se llama automáticamente al crear un objeto (inicializa atributos)
# self → referencia al objeto actual
# saludar → método que puede usar los atributos del objeto

# ------------------------------------------------------------
# 2. Crear un objeto (instanciar)
p1 = Persona("Ana", 25)
p2 = Persona("Víctor", 30)
# p1 y p2 son objetos distintos, aunque ambos usan la misma clase

# ------------------------------------------------------------
# 3. Llamar métodos del objeto
p1.saludar()  # Hola, mi nombre es Ana y tengo 25 años
p2.saludar()  # Hola, mi nombre es Víctor y tengo 30 años

# ------------------------------------------------------------
# 4. Acceder a atributos directamente
print(p1.nombre)  # Ana
print(p2.edad)    # 30

# ------------------------------------------------------------
# 5. Resumen rápido
# Concepto      Qué hace
# class         Define una clase (plantilla)
# __init__      Constructor, inicializa los atributos de un objeto
# self          Referencia al objeto actual
# Método        Función dentro de la clase que puede usar atributos del objeto
# Objeto        Instancia concreta de la clase

# ------------------------------------------------------------
# 6. Ejemplo completo con coche
class Coche:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        return f"Coche: {self.marca} {self.modelo}"

# Crear objetos
c1 = Coche("Toyota", "Corolla")
c2 = Coche("Ford", "Fiesta")

# Usar métodos
print(c1.descripcion())  # Coche: Toyota Corolla
print(c2.descripcion())  # Coche: Ford Fiesta

# ------------------------------------------------------------
# 7. Otro ejemplo de dos resultados con distinta forma de llegar a lo mismo

# Versión 1: atributos separados
class Person:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname

my_person = Person("Brais", "Moure")
print(my_person.name)      # Brais
print(my_person.surname)   # Moure

# Versión 2: atributo combinado
class PersonFullName:
    def __init__(self, name, surname):
        self.full_name = f"{name} {surname}"

my_person2 = PersonFullName("Brais", "Moure")
print(my_person2.full_name)  # Brais Moure
```

# Exceptions

```python
# ------------------------------------------------------------
# EXCEPCIONES EN PYTHON
# ------------------------------------------------------------
# Las excepciones son errores que ocurren mientras el programa se está ejecutando.
# Si no las controlas, tu programa se detiene y muestra un mensaje de error.

# Ejemplo de error sin manejo:
# print(10 / 0)
# 💥 Da error:
# ZeroDivisionError: division by zero

# ------------------------------------------------------------
# 1️⃣ Manejo básico con try y except
# Para evitar que tu programa se bloquee, usas un bloque try / except
try:
    print(10 / 0)
except:
    print("Ocurrió un error al dividir.")
# 🟢 Salida:
# Ocurrió un error al dividir.

# ------------------------------------------------------------
# 2️⃣ Especificar el tipo de error
# Puedes atrapar errores concretos, lo que es más limpio
try:
    print(10 / 0)
except ZeroDivisionError:
    print("No se puede dividir entre cero.")

# ------------------------------------------------------------
# 3️⃣ Usar else y finally
# else: se ejecuta si no hay errores.
# finally: se ejecuta siempre, haya error o no (ideal para cerrar archivos, conexiones, etc.)
try:
    resultado = 10 / 2
except ZeroDivisionError:
    print("Error: división entre cero.")
else:
    print("Todo salió bien:", resultado)
finally:
    print("Fin del bloque try/except.")
# 🟢 Salida:
# Todo salió bien: 5.0
# Fin del bloque try/except.

# ------------------------------------------------------------
# 4️⃣ Lanzar tus propias excepciones (raise)
# Puedes crear errores intencionalmente si algo no cumple una condición
edad = -5
if edad < 0:
    raise ValueError("La edad no puede ser negativa.")
# 💥 Esto detiene el programa con:
# ValueError: La edad no puede ser negativa.

# ------------------------------------------------------------
# 5️⃣ Resumen
# Palabra clave   Qué hace
# try             Bloque donde puede ocurrir un error
# except           Qué hacer si ocurre un error
# else            Se ejecuta si no hubo error
# finally         Se ejecuta siempre, pase lo que pase
# raise           Lanza un error manualmente
```

```python
# ------------------------------------------------------------
# 1️⃣ Qué es NumPy
# ------------------------------------------------------------
# NumPy es una librería fundamental para cálculo numérico en Python.
# Permite trabajar con arrays multidimensionales (más potentes que las listas).
# Está optimizada en C, por lo que es mucho más rápida que las estructuras nativas de Python para operaciones numéricas.

# Google Colab:
# NumPy ya viene preinstalado, solo hay que importarlo
import numpy as np

# Visual Studio / VS Code (Python local):
# Primero hay que instalar NumPy usando pip en la terminal:
# pip install numpy

# Luego se importa en tu script
import numpy as np

# ⚠️ Si usas entornos virtuales, asegúrate de instalar NumPy en el entorno correcto

# ------------------------------------------------------------
# 2️⃣ Creación de arrays
# ------------------------------------------------------------

# Array 1D
a = np.array([1, 2, 3])
print("Array 1D:", a)

# Array 2D
b = np.array([[1, 2], [3, 4]])
print("Array 2D:\n", b)

# Arrays especiales
zeros = np.zeros((3,3))
ones = np.ones((2,2))
arange = np.arange(0,10,2)
linspace = np.linspace(0,1,5)

print("Ceros:\n", zeros)
print("Unos:\n", ones)
print("Arange:", arange)
print("Linspace:", linspace)

# ------------------------------------------------------------
# 3️⃣ Propiedades de los arrays
# ------------------------------------------------------------
print("a.shape:", a.shape)  # Dimensiones del array
print("a.ndim:", a.ndim)    # Número de dimensiones
print("a.size:", a.size)    # Total de elementos
print("a.dtype:", a.dtype)  # Tipo de dato

# ------------------------------------------------------------
# 4️⃣ Indexación y slicing
# ------------------------------------------------------------
a = np.array([10, 20, 30, 40, 50])
print("Primer elemento:", a[0])
print("Subarray a[1:4]:", a[1:4])
print("Último elemento:", a[-1])

# Arrays 2D
b = np.array([[1,2,3],[4,5,6]])
print("Elemento fila 1, columna 2:", b[1,2])
print("Segunda columna:", b[:,1])

# ------------------------------------------------------------
# 5️⃣ Operaciones aritméticas
# ------------------------------------------------------------
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Suma:", a + b)
print("Multiplicación:", a * b)
print("Potencia:", a ** 2)

# Funciones universales
print("Raíz cuadrada:", np.sqrt(a))
print("Exponencial:", np.exp(a))
print("Media:", np.mean(a))
print("Máximo:", np.max(a))

# ------------------------------------------------------------
# 6️⃣ Operaciones con matrices
# ------------------------------------------------------------
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("Producto matricial np.dot(A,B):\n", np.dot(A,B))
print("Transpuesta A.T:\n", A.T)
print("Inversa np.linalg.inv(A):\n", np.linalg.inv(A))
print("Determinante np.linalg.det(A):", np.linalg.det(A))

# ------------------------------------------------------------
# 7️⃣ Boolean masking y filtros
# ------------------------------------------------------------
a = np.array([10, 20, 30, 40])
print("Elementos mayores a 20:", a[a > 20])

# ------------------------------------------------------------
# 8️⃣ Uso con datos reales
# ------------------------------------------------------------
# NumPy se usa como base de Pandas, Scikit-learn, Matplotlib.
# Es la base para cualquier trabajo en ciencia de datos o machine learning.
# Su fuerza está en manipular datos rápidamente y con poca memoria.
```
